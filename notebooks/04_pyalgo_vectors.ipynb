### Vectorization with NumPy

```python
sm = [2 * i for i in v]
```

```python
sm
```

```python
2 * v
```

```python
import numpy as np
```

```python
a = np.array(v)
```

```python
a
```

```python
a = np.arange(12).reshape((4, 3))
```

```python
a
```

```python
np.mean(a)
```
