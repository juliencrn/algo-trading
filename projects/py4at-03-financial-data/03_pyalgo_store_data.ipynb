### Using TsTables

```python
data = generate_sample_data(rows=2.5e6, cols=5, freq='1s').round(4)
```

```python
data.info()
```

```python
data
```

```python
import tstables
```

```python
import tables as tb
```

```python
class desc(tb.IsDescription):
    ''' Description of TsTables table structure.
    '''
    timestamp = tb.Int64Col(pos=0)
    No0 = tb.Float64Col(pos=1)
    No1 = tb.Float64Col(pos=2)
    No2 = tb.Float64Col(pos=3)
    No3 = tb.Float64Col(pos=4)
    No4 = tb.Float64Col(pos=5)
```

```python
h5 = tb.open_file('data.h5ts', 'a')
```

```python
ts = h5.create_ts('/', 'data3', desc)
```

```python
h5
```

```python
data
```

### Storing Data with SQLite3

```python
data = generate_sample_data(1e6, 5, '1min').round(4)
```

```python
data.info()
```

```python
import sqlite3 as sq3
```

```python
con = sq3.connect('data.sql')
```

```python
data.to_sql('/', con)
```

```python
query = 'SELECT * FROM data WHERE No1 > 105 and No2 < 108'
```

```python
res = con.execute(query).fetchall()
```

```python
res[:5]
```

```python
len(res)
```

```python
con.close()
```
